Various visual stimuli in pgl

```python
# library imports
import PIL.Image as Image

# Load PGL libraries and start a PGL window
from pgl import pgl as pglLibrary
pgl = pglLibrary()

# close any existing windows
pgl.cleanUp()

# open a new window
pgl.open(0,800,600)
pgl.visualAngle(57, 60,40)
```

Display an image

```python
# Load the image using PIL
imData = Image.open('/Users/justin/proj/pgl/testimage.jpg')

# display the image in the PGL window
pgl.imageDisplay(imData)
pgl.flush()
```

Raidal Checkerboard

```python
# Init a radial checkerboard stimulus
radialCheckerboard = pgl.radialCheckerboard(pgl,type='sliding',theta=0,innerRadius=0,outerRadius=6,radialWidth=360,checkRadialWidth=7.5,temporalSquareWave=False)

# print the stimulus object
print(radialCheckerboard)

# display duration in second
duration = 2
startTime = pgl.getSecs()

# display the stimulus
while pgl.getSecs() - startTime < duration:
    radialCheckerboard.display()
    pgl.flush()
```

Dots

```python
# Init a random dot kinematogram stimulus
rdk = pgl.randomDots(width=10, height=10, noiseType='movshon')

# print the stimulus object
print(rdk)

# display duration in second
duration = 2
startTime = pgl.getSecs()

# display the stimulus
while pgl.getSecs() - startTime < duration:
    rdk.display(direction=181, coherence=1, speed=7.0)
    pgl.flush()
    
```

Checkerboard stimulus

```python
# initialize the checkerboard stimulus
checkerboard = pgl.checkerboard(type='flickering',checkWidth=1, checkHeight=1,temporalFrequency=5)

# display duration in second
duration = 2
startTime = pgl.getSecs()

# display the stimulus
while pgl.getSecs() - startTime < duration:
    checkerboard.display()
    pgl.flush()
```

Bars

```python
# initialize bar stimulus
bar = pgl.bar(width=1, speed=5, dir=45)

# display duration in second
duration = 10
startTime = pgl.getSecs()

# display the stimulus
while pgl.getSecs() - startTime < duration:
    bar.display()
    pgl.flush()
    
# change direction
bar.dir = 90

# display the stimulus
startTime = pgl.getSecs()
while pgl.getSecs() - startTime < duration:
    bar.display()
    pgl.flush()
```

Draw text

```python

# Draw text in center of the screen
pgl.text("hello world",x=0,y=0)
pgl.flush()
```

Specify font, size and color

```python
# Draw text in center of the screen
pgl.text("hello world",x=0,y=0,fontName = "Futura",fontSize=50,color=(0,1,1,1))
pgl.flush()
```

You can align the text in various ways

```python
# align text to left
pgl.text("align left",x=1,y=0,xAlign=-1)

# align text to right
pgl.text("align right",x=-1,y=0,xAlign=1)

# align text to top
pgl.text("align top",x=0,y=-1,yAlign=-1)

# align text to bottom
pgl.text("align bottom",x=0,y=1,yAlign=1)

# flush the display
pgl.flush()
```

Instead of an x and y position, you can specify line number to display, where + values are from top of screen and - values are from bottom of screen. Subsequent lines are automatically placed below the last line you draw, and if it goes below the bottom of the screen, it will wrap to the top

```python
# draw a line of text 2 lines from the bottom
pgl.text("Hello again!",line=-2)

# draw two more lines of text, note that the last line will
# wrap to the top of the screen
pgl.text("next line")
pgl.text("and the next line")

# flush the display
pgl.flush()
```

If you want to display the same line of text over and over again, you may want to create the text once (it is faster in your display loop not to have to regenerate the text image)

```python
# create a text image
textImage = pgl.textCreate("A line of text to draw many times, so create in advance")

# now I can draw the text a few times 
# note that I could do this across multiple flushes rather
# than on the same image (which would probably be more typical)
for iRepeat in range(10):
    pgl.text(textImage)

# flush the display
pgl.flush()
```

Draw an arc

```python
# draw a filled arc
pgl.arc(0,0,innerRadius=0,outerRadius=10,color=(0,1,0.7,1))

# and flush the display
pgl.flush()
```

Draw a grating. 

```python
# initialize the grating
grating = pgl.grating(spatialFrequency=0.1,phase=0, contrast=1,temporalFrequency=4.0,direction=-1)

# clear the screen to gray
pgl.clearScreen((0.5, 0.5, 0.5))
pgl.flush()

# display duration in second
duration = 2
startTime = pgl.getSecs()

# display the grating stimulus
while pgl.getSecs() - startTime < duration:
    grating.display()
    pgl.flush()
```
