Movie display tutorial

```python
# setup autoreload
%load_ext autoreload
%autoreload 2

# Load PGL libraries and start a PGL window
from pgl import pgl as pglLibrary
pgl = pglLibrary()

# close any existing windows
pgl.cleanUp()

# open a new window
pgl.open(0,800,600)
pgl.visualAngle(57, 60,40)
```

Display the movie and show the frame statistics

```python
# start by flushing the screen to get the display loop going (this would normally done in the expt code)
pgl.flush(); pgl.flush(); pgl.flush(); pgl.flush();

# play the movie
m.play()
```
