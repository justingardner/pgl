```python
import subprocess, json

def getGPUInfoJSON():
    out = subprocess.run(
        ["system_profiler", "SPDisplaysDataType", "-json"],
        capture_output=True, text=True, check=True
    ).stdout
    return json.loads(out)

def getDisplayIdentityJSON():
    out = subprocess.run(
        ["system_profiler", "SPDisplaysDataType", "-json"],
        capture_output=True, text=True, check=True
    ).stdout
    data = json.loads(out)

    displays = []
    for gpu in data.get("SPDisplaysDataType", []):
        for disp in gpu.get("spdisplays_ndrvs", []):
            displays.append({
                'name':           disp.get('_name'),
                'displayId':      disp.get('_spdisplays_displayID'),
                'vendorId':       hexToDec(disp.get('_spdisplays_display-vendor-id')),
                'productId':      hexToDec(disp.get('_spdisplays_display-product-id')),
                'serial':         hexToDec(disp.get('_spdisplays_display-serial-number')),
                'displayType':    disp.get('spdisplays_display_type'),
                'connectionType': disp.get('spdisplays_connection_type'),
                'pixels':         disp.get('_spdisplays_pixels'),
                'resolution':     disp.get('_spdisplays_resolution'),
                'isMain':         disp.get('spdisplays_main') == 'spdisplays_yes',
            })
            print(displays[-1])
    return displays
def hexToDec(value):
    """Convert a hex string like 'a04f' to a decimal int, safely.

    Returns None if value is None or not valid hex.
    """
    if value is None:
        return None
    try:
        return int(str(value), 16)
    except ValueError:
        return None
displays = getDisplayIdentityJSON()
```

```python
import sys
from PySide6.QtWidgets import (
    QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout,
    QLabel, QComboBox, QListWidget, QRadioButton, QPushButton,
    QGroupBox, QButtonGroup, QStatusBar
)
from PySide6.QtGui import QAction

class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("Selection Tool")
        self.resize(500, 400)

        # --- Menu bar ---
        menu = self.menuBar()
        file_menu = menu.addMenu("&File")
        file_menu.addAction(QAction("Export...", self, triggered=self.on_export))
        file_menu.addSeparator()
        file_menu.addAction(QAction("Exit", self, triggered=self.close))

        help_menu = menu.addMenu("&Help")
        help_menu.addAction(QAction("About", self, triggered=self.on_about))

        # --- Central widget ---
        central = QWidget()
        layout = QVBoxLayout(central)
        layout.setContentsMargins(20, 20, 20, 20)
        layout.setSpacing(14)

        layout.addWidget(QLabel("<b>Choose a category:</b>"))
        self.combo = QComboBox()
        self.combo.addItems(["Alpha", "Beta", "Gamma", "Delta"])
        layout.addWidget(self.combo)

        layout.addWidget(QLabel("<b>Pick items:</b>"))
        self.list_widget = QListWidget()
        self.list_widget.addItems([f"Item {i}" for i in range(1, 8)])
        self.list_widget.setSelectionMode(QListWidget.SelectionMode.ExtendedSelection)
        layout.addWidget(self.list_widget)

        group_box = QGroupBox("Mode")
        group_layout = QHBoxLayout()
        self.radio_group = QButtonGroup(self)
        for i, mode in enumerate(["Fast", "Balanced", "Thorough"]):
            rb = QRadioButton(mode)
            if i == 1:
                rb.setChecked(True)
            self.radio_group.addButton(rb)
            group_layout.addWidget(rb)
        group_box.setLayout(group_layout)
        layout.addWidget(group_box)

        run_btn = QPushButton("Run")
        run_btn.clicked.connect(self.on_run)
        layout.addWidget(run_btn)

        self.setCentralWidget(central)
        self.setStatusBar(QStatusBar())

        self.setStyleSheet("""
            QMainWindow { background-color: #f5f5f7; }
            QLabel { font-size: 13px; }
            QPushButton {
                background-color: #2563eb; color: white;
                border-radius: 6px; padding: 8px 16px; font-weight: 600;
            }
            QPushButton:hover { background-color: #1d4ed8; }
            QComboBox, QListWidget {
                border: 1px solid #d1d5db; border-radius: 6px; padding: 4px;
                background: white;
            }
            QGroupBox {
                border: 1px solid #d1d5db; border-radius: 6px;
                margin-top: 8px; padding-top: 10px; font-weight: 600;
            }
        """)

    def on_run(self):
        selected = [i.text() for i in self.list_widget.selectedItems()]
        self.statusBar().showMessage(
            f"Category: {self.combo.currentText()} | Selected: {selected}"
        )

    def on_export(self):
        self.statusBar().showMessage("Export clicked (stub)")

    def on_about(self):
        self.statusBar().showMessage("Selection Tool v1.0")


app = QApplication.instance() or QApplication(sys.argv)
window = MainWindow()
window.show()
app.exec()
```

```python
import sys
import numpy as np
from PySide6.QtWidgets import (
    QApplication, QMainWindow, QWidget, QVBoxLayout, QHBoxLayout,
    QLabel, QComboBox, QPushButton
)
from matplotlib.backends.backend_qtagg import FigureCanvasQTAgg as FigureCanvas
from matplotlib.figure import Figure


class MainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.setWindowTitle("GUI with Embedded Plot")
        self.resize(700, 500)

        central = QWidget()
        layout = QVBoxLayout(central)
        layout.setContentsMargins(20, 20, 20, 20)
        layout.setSpacing(12)

        # --- Controls ---
        controls = QHBoxLayout()
        controls.addWidget(QLabel("Function:"))
        self.combo = QComboBox()
        self.combo.addItems(["sin", "cos", "sin*cos", "random walk"])
        self.combo.currentTextChanged.connect(self.update_plot)
        controls.addWidget(self.combo)

        refresh_btn = QPushButton("Refresh")
        refresh_btn.clicked.connect(self.update_plot)
        controls.addWidget(refresh_btn)
        controls.addStretch()
        layout.addLayout(controls)

        # --- Matplotlib canvas ---
        self.figure = Figure(figsize=(5, 4))
        self.canvas = FigureCanvas(self.figure)
        self.ax = self.figure.add_subplot(111)
        layout.addWidget(self.canvas)

        self.setCentralWidget(central)

        self.setStyleSheet("""
            QMainWindow { background-color: #f5f5f7; }
            QPushButton {
                background-color: #2563eb; color: white;
                border-radius: 6px; padding: 6px 14px; font-weight: 600;
            }
            QPushButton:hover { background-color: #1d4ed8; }
            QComboBox {
                border: 1px solid #d1d5db; border-radius: 6px;
                padding: 4px 8px; background: white;
            }
        """)

        self.update_plot()

    def update_plot(self):
        x = np.linspace(0, 10, 300)
        choice = self.combo.currentText()

        if choice == "sin":
            y = np.sin(x)
        elif choice == "cos":
            y = np.cos(x)
        elif choice == "sin*cos":
            y = np.sin(x) * np.cos(x)
        else:  # random walk
            y = np.cumsum(np.random.randn(300))

        self.ax.clear()
        self.ax.plot(x, y, color="#2563eb", linewidth=2)
        self.ax.set_title(choice)
        self.ax.grid(alpha=0.3)
        self.canvas.draw()


app = QApplication.instance() or QApplication(sys.argv)
window = MainWindow()
window.show()
app.exec()
```

```python
import sys
from dataclasses import dataclass, fields, is_dataclass
from typing import get_type_hints, get_args, Literal
from enum import Enum

from PySide6.QtWidgets import (
    QApplication, QDialog, QVBoxLayout, QFormLayout,
    QCheckBox, QComboBox, QPushButton, QLabel, QDialogButtonBox
)


# ---- 1. Define your options as a dataclass ----
class Method(Enum):
    FAST = "Fast"
    BALANCED = "Balanced"
    THOROUGH = "Thorough"


@dataclass
class RunConfig:
    normalize: bool = True
    include_outliers: bool = False
    method: Method = Method.BALANCED
    scale: Literal["linear", "log", "sqrt"] = "linear"


# ---- 2. Generic form generator ----
class ConfigDialog(QDialog):
    def __init__(self, config_instance):
        super().__init__()
        self.setWindowTitle("Configure Run")
        self.config = config_instance
        self.widgets = {}

        layout = QVBoxLayout(self)
        layout.addWidget(QLabel(f"<b>{type(config_instance).__name__}</b>"))

        form = QFormLayout()
        hints = get_type_hints(type(config_instance))

        for f in fields(config_instance):
            name = f.name
            value = getattr(config_instance, name)
            ftype = hints[name]

            if ftype is bool:
                w = QCheckBox()
                w.setChecked(value)

            elif isinstance(ftype, type) and issubclass(ftype, Enum):
                w = QComboBox()
                for member in ftype:
                    w.addItem(member.value, member)
                w.setCurrentIndex(w.findData(value))

            elif get_args(ftype):  # Literal["a", "b", "c"]
                w = QComboBox()
                for option in get_args(ftype):
                    w.addItem(str(option), option)
                w.setCurrentIndex(w.findData(value))

            else:
                continue  # unsupported type, skip

            self.widgets[name] = w
            form.addRow(name.replace("_", " ").title() + ":", w)

        layout.addLayout(form)

        buttons = QDialogButtonBox(
            QDialogButtonBox.StandardButton.Ok | QDialogButtonBox.StandardButton.Cancel
        )
        buttons.accepted.connect(self.accept)
        buttons.rejected.connect(self.reject)
        layout.addWidget(buttons)

        self.setStyleSheet("""
            QDialog { background-color: #f5f5f7; }
            QPushButton {
                background-color: #2563eb; color: white;
                border-radius: 6px; padding: 6px 14px; font-weight: 600;
            }
            QComboBox { border: 1px solid #d1d5db; border-radius: 6px; padding: 4px; background: white; }
        """)

    def get_result(self):
        """Read widget values back into a new instance of the dataclass."""
        values = {}
        for name, w in self.widgets.items():
            if isinstance(w, QCheckBox):
                values[name] = w.isChecked()
            elif isinstance(w, QComboBox):
                values[name] = w.currentData()
        return type(self.config)(**values)


# ---- 3. Run it and get the result back ----
app = QApplication.instance() or QApplication(sys.argv)

dialog = ConfigDialog(RunConfig())
result_config = None
if dialog.exec() == QDialog.DialogCode.Accepted:
    result_config = dialog.get_result()

print(result_config)
```
